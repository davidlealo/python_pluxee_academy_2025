```python
encuesta = pd.read_csv('encuesta.csv', sep=",")
display(encuesta)
```

```python
print(type(encuesta))
```

```python
encuestaString = encuesta.to_string(index=False)
```

```python
print(type(encuestaString))
```

```python
# Hacer la consulta sobre la variable encuestaString

response = openai.responses.create(
    model="gpt-4.1-nano",
    input="Qué entiendes de: " + encuestaString
)

print(response.output_text)
```

```python
# Hacer la consulta sobre la variable encuestaString con más contexto

response = openai.responses.create(
    model="gpt-4o",
    input="Qué entiendes de lo siguiente:  " + encuestaString + " Sabiendo que la columna sí/no refleja si las personas comprendieron la clase. La siguiente columna es alguna observación que escribieron en una encuesta. Dame recomendaciones para mejorar mi clase"
)

print(response.output_text)
```

```python
import openai

response = openai.images.generate(
    model="dall-e-3",  # O puedes usar "dall-e-2"
    prompt="Una infografía que represente los comentarios de una encuesta sobre si los estudiantes entendieron una clase. Algunos comentarios dicen que necesitan más ejemplos, otros mencionan problemas de conexión, y otros agradecen la paciencia del profesor.",
    size="1024x1024",  # Tamaños posibles: 256x256, 512x512, 1024x1024
    n=1
)

image_url = response.data[0].url
print(image_url)
```

```python
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO

# Descargar la imagen
img_data = requests.get(image_url).content
img = Image.open(BytesIO(img_data))

# Mostrarla con matplotlib
plt.imshow(img)
plt.axis('off')  # Oculta los ejes
plt.show()
```
